```python
from scripts.utils import get_creds

from googleapiclient.discovery import build
```

```python
creds = get_creds()
service = build("calendar", "v3", credentials=creds)
```

```python
primary_calendar = (service
                    .calendars()
                    .get(calendarId="primary")
                    .execute()
                    )
```

```python
primary_calendar
```

```python
fields = "items"
events = service.events().list(calendarId="primary", fields=fields).execute()
```

```python
events
```

```python
events["items"][0]["created"]
```

```python
birthdays_to_delete = [event for event in events["items"] if "summary" in event and event["summary"].startswith("Birthday of")]
```

```python
birthdays_to_delete = [event for event in events["items"] if "description" in event and event["description"].startswith("Birthday")]
```

```python
from datetime import datetime

def is_on_date(event, month, day) -> bool:
    date = datetime.strptime(event["created"][:-5], "%Y-%m-%dT%H:%M:%S")
    return date.month == month and date.day == day
```

```python
birthdays_to_delete = [event for event in events["items"] if is_on_date(event, 3, 27)]
```

```python
birthdays_to_delete
```

```python
from tqdm import tqdm

for birthday in tqdm(birthdays_to_delete):
    service.events().delete(calendarId="primary", eventId=birthday["id"]).execute()
```
